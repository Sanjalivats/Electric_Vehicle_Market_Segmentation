#**Label encoding**

```python
from sklearn.preprocessing import LabelEncoder
def label_encoder(data,column):
    label_encoder=LabelEncoder()
    return label_encoder.fit_transform(data[column].astype(str))
```

#**K Means clustering Model**

###**Checking for Variance Inflation Factor**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def vif(x):
    vif=pd.DataFrame()
    vif['variables']=x.columns
    vif['VIF']=[variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
    return(vif)
```

```python
x=df.iloc[:,:-1]
```

```python
vif(x)
```

##**Implementing PCA**

```python
from sklearn.decomposition import PCA

# Performing PCA
pca = PCA(n_components=4)
pca_data = pca.fit_transform(df)
df2 = pd.DataFrame(pca_data,columns= ['PC1','PC2','PC3','PC4'])
df2.head()
```

##**Finding number of clusters using Elbow Method**

```python
from sklearn.cluster import KMeans

# Determine the number of clusters using the Elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(pca_data)
    wcss.append(kmeans.inertia_)

# Plot the Elbow curve
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.grid()
plt.show()
```

##**Applying the KMeans Clustering algorithm**

```python
# Perform K-means clustering algo. with the optimal number of clusters
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
kmeans.fit(pca_data)
```

```python
# Add the cluster labels to the dataset
df['Cluster'] = kmeans.labels_

# Define a custom color and size
color_map = 'cool'  # Change the color map as desired
marker_size = 50  # Change the marker size as desired

# Visualize the clusters
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=df['Cluster'], cmap=color_map, s=marker_size)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('EV Market Segmentation - Cluster Analysis')
plt.colorbar(label='Cluster')
plt.grid()

plt.show()
```

##**Making Predictions**

```python
kmeans_predict=KMeans(n_clusters=4,random_state=123)
clusters=kmeans_predict.fit_predict(df)
df_copy["Clusters"]=clusters
```
